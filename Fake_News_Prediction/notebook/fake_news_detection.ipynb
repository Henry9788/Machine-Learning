```python
import pandas as pd
import re
import pickle
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.tree import DecisionTreeClassifier
```

```python
true = pd.read_csv('..\Dataset\True.csv')
df_true = pd.DataFrame(true)
df_true.head()
```

```python
fake = pd.read_csv('..\Dataset\Fake.csv')
df_fake = pd.DataFrame(fake)
df_fake.head()
```

```python

df_true['label'] = 1
df_fake['label'] = 0
```

```python
df_true.head()
```

```python
df_news = pd.concat([df_fake,df_true],axis=0)
df_news.head()
```

```python
df_news.tail()
```

```python
df_news.isnull().sum()
```

```python
df_news = df_news.drop(['title','subject','date'],axis=1)
df_news.head()
```

```python
df_news = df_news.sample(frac=1)    #Reshuffling
df_news.head()
```

```python
df_news.reset_index(inplace=True)
df_news.head()
```

```python
df_news.drop(['index'],axis=1)
```

```python
# feature extraction
def fnTextOptim(text):
    text = text.lower() # convert lower case
    text = re.sub(r'https?://\S+|www\.\S+','',text) #remove urls
    text = re.sub(r'<.*?>','',text) #remove urls
    text = re.sub(r'[^\w\s]','',text)   # remove punctuation
    text = re.sub(r'\d','',text)    #remove digits
    text = re.sub(r'\n',' ',text)    #remove new line
    
    return text
```

```python
df_news['text'] = df_news['text'].apply(fnTextOptim)
df_news
```

```python
x = df_news['text']
y = df_news['label']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
model = DecisionTreeClassifier()
model = model.fit(xv_train,y_train)
```

```python
y_pred = model.predict(xv_test)
model.score(xv_test,y_test)
```

```python
with open('fake_news_detect_model.pkl','wb') as file:
    pickle.dump(model,file)
    
with open('tfidf_vectorizer.pkl','wb') as file:
    pickle.dump(vectorization,file)
```

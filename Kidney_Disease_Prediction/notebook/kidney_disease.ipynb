```python
import pandas as pd
```

```python
data = pd.read_csv("..\Dataset\kidney_disease.csv")
df_data = pd.DataFrame(data)
df_data
```

```python
df_data.drop("id",axis=1,inplace=True)

df_data.columns = ['age','blood_presure','specific_gravity','albumin','sugar','red_bled_cells','pus_cel','pus_cel_clumps','bacteria',
                   'blood_lucose_random','blood_urea','serum_creatinine','sodium','pottasium','hemoglobin','packed_cell_volume','white_blood_cell_count',
                   'red_blood_cell','hypertension','diabetes_mellius','coromary_artery_disease','appetite','peda_edema','anemia','class']
df_data
```

```python
df_data.info()
```

```python
df_data.isnull().sum()
```

```python
numeric_column = ['packed_cell_volume','white_blood_cell_count','red_blood_cell']
for col in numeric_column:
    print(f"{col} : {df_data[col].dtype}")
```

```python
def convert_to_numeric(df_data,column):
    df_data[column] = pd.to_numeric(df_data[column],errors='coerce')
    
for col in numeric_column:
    convert_to_numeric(df_data,col)
    print(f"{col} : {df_data[col].dtype}")
df_data
```

```python
def mean_value_computation(df_data,column):
    mean_val = df_data[column].mean()
    df_data[column].fillna(value=mean_val, inplace=True)
    
def mode_value_imputation(df_data,column):
    mode_val = df_data[column].mode()[0]
    df_data[column].fillna(value=mode_val, inplace=True)
```

```python
numeric_col = [col for col in df_data.columns if df_data[col].dtype != 'object']
text_col = [col for col in df_data.columns if df_data[col].dtype == 'object']
```

```python
for i in numeric_col:
    mean_value_computation(df_data,i)
    
for j in text_col:
    mode_value_imputation(df_data,j)
    print(f"{j} : {df_data[j].unique()}")
```

```python
df_data['diabetes_mellius'] = df_data['diabetes_mellius'].replace(to_replace={" yes":"yes","\tno":"no","\tyes":"yes"})
df_data['coromary_artery_disease'] = df_data['coromary_artery_disease'].replace(to_replace={"\tno":"no"})
df_data['class'] = df_data['class'].replace(to_replace={"ckd\t":"ckd"})
df_data
for j in text_col:
    print(f"{j} : {df_data[j].unique()}")
```

```python
df_data['red_bled_cells'] = df_data['red_bled_cells'].map({"normal":1,"abnormal":0})
df_data["pus_cel"] = df_data["pus_cel"].map({'normal':'1','abnormal':'0'})
df_data["pus_cel_clumps"] = df_data["pus_cel_clumps"].map({'present':'1','notpresent':'0'})
df_data["bacteria"] = df_data["bacteria"].map({'present':'1','notpresent':'0'})
df_data["hypertension"] = df_data["hypertension"].map({'yes':'1','no':'0'})
df_data["diabetes_mellius"] = df_data["diabetes_mellius"].map({'yes':1,'no':0})
df_data["coromary_artery_disease"] = df_data["coromary_artery_disease"].map({'yes':'1','no':'0'})
df_data["appetite"] = df_data["appetite"].map({'good':'1','poor':'0'})
df_data["peda_edema"] = df_data["peda_edema"].map({'yes':'1','no':'0'})
df_data["anemia"] = df_data["anemia"].map({'yes':'1','no':'0'})
df_data["class"] = df_data["class"].map({'ckd':'1','notckd':'0'})
df_data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
corr = df_data.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr,annot=True,linewidths=0.5)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,mean_squared_error,r2_score,mean_absolute_error
```

```python
x = df_data.drop('class',axis=1)
y = df_data['class']

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.25,random_state=42)
```

```python
models =[]
models.append(("Decision Tree Classifier",DecisionTreeClassifier()))
models.append(("GaussianNB",GaussianNB()))
models.append(("Random Forest Classifier",RandomForestClassifier()))
models.append(("KNeighbors Classifier",KNeighborsClassifier()))
models.append(("SVC",SVC()))

for name,model in models:
    model_fit = model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    print("")
    print(name)
    print(f"accuracy_score:{accuracy_score(y_pred,y_test)}")
    print(f"mean_squared_error:{mean_squared_error(y_pred,y_test)}")
    print(f"mean_absolute_error:{mean_absolute_error(y_pred,y_test)}")
    print(f"r2_score:{r2_score(y_pred,y_test)}")
```
